## Clean Data

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import re
import gc

def cleanString(String):
    ps = PorterStemmer()
    String = re.sub('[^a-zA-Z]', ' ', String).strip()
    String = String.replace('  ', ' ')
    String = String.lower()
    String = String.split()
    String = [ps.stem(word) for word in String if not word in stopwords.words('english')]
    String = ' '.join(String)
    return String
```

```python
def preprocess(inp_df):
    df = copy.deepcopy(inp_df)
    df['content'] = df['content'].apply(lambda string: cleanString(string.upper()))
    df['title'] = df['title'].apply(lambda string: cleanString(string.upper()))
    return df
```

```python
def sbert_embedder(model, df, NonStdCol="NonStdCol"):
    features = df[NonStdCol].apply(lambda x: list(model.encode(x))).tolist()
    print(f"Generated features matrix shape: {len(features)}")
    feature_col_names = ["SBERT_" + NonStdCol + "_feat_"+str(id+1) for id in range(len(features[0]))]
    features_df = pd.DataFrame(data=features, columns=feature_col_names)
    print("Features df:")
    display(features_df.head())
    df[feature_col_names] = features_df
    return df
```
